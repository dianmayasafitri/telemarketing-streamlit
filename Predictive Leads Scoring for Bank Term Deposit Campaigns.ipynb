# **5. Evaluation Metrics**

Evaluation metrics assess how well the predictive model distinguishes between customers who are likely to subscribe and those who are not. 
Performance evaluation relies on the Confusion Matrix, which consists of:  
-   True Positive (TP): predicted “yes” and actually “yes”.
-   True Negative (TN): predicted “no” and actually “no”.
-   False Positive (FP): predicted “yes” but actually “no”.
-   False Negative (FN): predicted “no” but actually “yes”.

Various evaluation metrics are derived from these four components:

### **Precision**
Formula (plain text):  
Precision = TP / (TP + FP)

### **Recall**
Formula (plain text):  
Recall = TP / (TP + FN)

### **F-Score (general formula)**
Formula (plain text):  
F_beta = (1 + beta^2) * (Precision * Recall) / ((beta^2 * Precision) + Recall)

---

## **False Positive (FP) and False Negative (FN) in the Context of Deposit Campaigns**

### False Positive (FP)  
The model predicts a customer will subscribe, but they are not actually interested.
Impact:
- Increased operational costs.
- Wasted agent time.
- Higher Cost per Acquisition.

### False Negative (FN)  
The model predicts a customer will not subscribe, but they are actually interested.
Impact:
- Lost deposit revenue.
- Lower overall conversion rates.
- High-value leads are missed.


### Model Evaluation Priority

False Negatives (FN) is prioritized to ensure that no potential customers are missed by the prediction system. A False Negative occurs when the model predicts that a customer will not place funds in a term deposit, when in fact they are interested. This type of error has a direct impact on lost revenue opportunities and lowers the campaign’s conversion rate.

Therefore, the primary metric we focus on is Recall, which measures the model’s ability to capture all customers who are truly interested. A high Recall indicates that the model successfully minimizes False Negatives, allowing more high-value prospects to be contacted by telemarketing agents. This approach ensures that the campaign runs more effectively, sales opportunities increase, and revenue potential is not wasted.

---

## **Primary Metric Selection: Recall**

In the context of term deposit telemarketing campaigns, every False Negative (FN) — a customer who is actually interested in subscribing but is predicted as not interested by the model — results in lost conversion opportunities and potential revenue. False Negatives cause agents to miss high-value customers who should have been contacted, thereby reducing campaign effectiveness. Therefore, as a Data Science team, we prioritize reducing FN as part of our model optimization strategy.

To achieve this objective, the primary metric used is Recall, which represents the proportion of truly interested customers that are successfully captured by the model.

Formula (plain text):
Recall = True Positive / (True Positive + False Negative)

By using Recall as the main metric, the model is optimized to minimize the number of False Negatives, ensuring that fewer potential customers are overlooked. This approach ensures that the resulting prospect list is more comprehensive, increases conversion opportunities, and maximizes the overall impact of the telemarketing campaign.

## 2.1. Mendefinisikan Fitur Numerikal dan Kategorikal

```python
# Select features:

candidate_features = [
    'age','job','marital','education','default','housing','loan',
    'contact','month','day_of_week','campaign','pdays','previous','poutcome',
    'emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed'
    ]

# Filter columns
use_features = [c for c in candidate_features if c in df.columns]
if 'duration' in use_features:
    use_features.remove('duration')
```

```python
# Education order (custom, required for ordinal encoder)”
education_order = [
    'illiterate',
    'basic.4y',
    'basic.6y',
    'basic.9y',
    'high.school',
    'professional.course',
    'university.degree',
    'unknown'
]

# Mapping features
ordinal_features = ['education']
force_categorical = [
    'job', 'marital', 'default', 'housing', 'loan',
    'contact', 'month', 'day_of_week', 'poutcome', 'generation'
]

# Separate numerical features
numeric_features = [
    c for c in use_features
    if (np.issubdtype(df[c].dtype, np.number))
       and c not in force_categorical
       and c not in ordinal_features
]

# Nominal categorical features
categorical_features = [
    c for c in use_features
    if c not in numeric_features
       and c not in ordinal_features
]

print("\nNumeric features:", numeric_features)
print("Categorical features:", categorical_features)
print("Ordinal features:", ordinal_features)
```

## 2.3. Preprocessing Pipeline 

```python
# 1. Numerical Pipeline
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler())
])

# 2. Ordinal Pipeline (education)
ordinal_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder(categories=[education_order]))
])

# 3. Nominal Categorical Pipeline — Binary Encoder
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('binary', ce.BinaryEncoder())
])

# 4. Combine everything into a single ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('ord', ordinal_transformer, ordinal_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='drop'
)
```

## 3.2. Cross-Validation with SMOTE

```python
## 3.2. Cross-Validation with SMOTE
scoring = {
    'f2': make_scorer(fbeta_score, beta=2),
    'recall': make_scorer(recall_score),
    'precision': make_scorer(precision_score),
    'accuracy': make_scorer(accuracy_score)
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

def model_cv_results(models, preprocessor):
    rows = []
    
    for name, model in models.items():
        print(f"CV running: {name}...")

        pipeline = ImbPipeline([
            ('preprocessor', preprocessor),
            ('smote', SMOTE(random_state=42)),
            ('classifier', model)
        ])

        scores = cross_validate(
            pipeline, X_train, y_train,
            scoring=scoring,
            cv=cv,
            n_jobs=-1,
            return_train_score=False
        )

        rows.append({
            "Model": name,
            "F2_mean": scores['test_f2'].mean(),
            "Recall_mean": scores['test_recall'].mean(),
            "Precision_mean": scores['test_precision'].mean(),
            "Accuracy_mean": scores['test_accuracy'].mean()
        })

    return pd.DataFrame(rows).sort_values("Recall_mean", ascending=False)

cv_leaderboard = model_cv_results(models, preprocessor).round(3)
cv_leaderboard
```

By systematically tuning the hyperparameters, the model is optimally evaluated in identifying customers who are truly interested but at risk of being missed by the model. This approach ensures that the final model delivers the best possible performance in minimizing FN, which is the top priority in the term deposit telemarketing campaign.

# 4. Cost-Benefit Calculator

Run Cost–Benefit Analysis

```python
def cost_benefit_analysis(
    y_true, 
    y_pred, 
    benefit_TP=0, 
    cost_FN=0,          # Focus on False Negatives (FN)
    cost_FP=0, 
    benefit_TN=0
):
    TP = np.sum((y_true == 1) & (y_pred == 1))
    FP = np.sum((y_true == 0) & (y_pred == 1))
    FN = np.sum((y_true == 1) & (y_pred == 0))
    TN = np.sum((y_true == 0) & (y_pred == 0))

    # Primary focus: False Negatives (FN)
    Cost_FN_Total = FN * cost_FN

    results = {
        "TP": int(TP),
        "FP": int(FP),
        "FN": int(FN),
        "TN": int(TN),

        # Primary focus:  FN
        "Operational_Cost_FN": int(Cost_FN_Total),
        "Cost_FN_Total": int(Cost_FN_Total),

        # False Positive (FP) costs are still calculated but not the main focus
        "Cost_FP_Total": int(FP * cost_FP),

        # Benefit
        "Benefit_TP_Total": int(TP * benefit_TP),
        "Benefit_TN_Total": int(TN * benefit_TN),

        # Net impact with a large penalty for FN
        "Net_Impact": int(
            (TP * benefit_TP + TN * benefit_TN)
            - (Cost_FN_Total + FP * cost_FP)
        )
    }
    
    return results
```
